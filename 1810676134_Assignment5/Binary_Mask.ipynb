```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def main():
    rgb=plt.imread("/home/shahadat/Desktop/DIP/b.jpeg")
    plt.figure(figsize=(50,100))
    plt.subplot(3,1,1)
    plt.title("RGB",fontsize=150)
    plt.imshow(rgb)

    grayscale=cv2.cvtColor(rgb,cv2.COLOR_RGB2GRAY)
    plt.subplot(3,1,2)
    plt.title("Grayscale",fontsize=150)
    plt.imshow(grayscale,cmap="gray")

    zero_mask=np.zeros(grayscale.shape ,dtype=np.uint8)
    zero_mask=cv2.circle(zero_mask,(70,60),50,(255,255,255),-1)

    mask_img=cv2.bitwise_and(grayscale,zero_mask)

    plt.subplot(3,1,3)
    plt.title("Binary_Mask_Img",fontsize=150)
    plt.imshow(mask_img,cmap='gray')

    plt.savefig("Binary_Mask.jpeg")
if __name__=='__main__':
    main()
```
