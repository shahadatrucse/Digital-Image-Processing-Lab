```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def main():
    img=plt.imread("/home/shahadat/Desktop/DIP/b.jpeg")
    plt.figure(figsize=(50,100))
    plt.subplot(5,2,1)
    plt.title("RGB",fontsize=150)
    plt.imshow(img)
    grayscale=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    plt.subplot(5,2,2)
    plt.title("Grayscale",fontsize=150)
    plt.imshow(grayscale,cmap='gray')
    w,h=grayscale.shape

    zeros=np.zeros((8,w,h))
    dataset=[128,64,32,16,8,4,2,1]

    for i in range(0,8):
        for j in range(0,w):
            for k in range(0,h):
                zeros[i][j][k]=grayscale[j][k] & dataset[i]

    pos=3
    m=7
    for n in range(0,8):
        plt.subplot(5,2,pos)
        plt.title(str(m+1)+"th Plane",fontsize=150)
        plt.imshow(zeros[n],cmap="gray")
        pos=pos+1
        m-=1
    plt.savefig("Bit_Slicing.jpeg")

if __name__=='__main__':
    main()
```
