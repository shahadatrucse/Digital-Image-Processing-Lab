```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def main(): 
    rgb=plt.imread("/home/shahadat/Desktop/DIP/sobel1.jpg")
    plt.figure(figsize=(50,100))
    plt.subplot(5,1,1)
    plt.title("RGB",fontsize=150)
    plt.imshow(rgb)

    grayscale=cv2.cvtColor(rgb,cv2.COLOR_RGB2GRAY)
    plt.subplot(5,1,2)
    plt.title("Grayscale",fontsize=150)
    plt.imshow(grayscale,cmap="gray")

    laplacian=np.array([[0,1,0],[1,-4,1],[0,1,0]])
    laplacian_img=cv2.filter2D(grayscale,-1,laplacian)
    plt.subplot(5,1,3)
    plt.title("Laplacian Filter",fontsize=150)
    plt.imshow(laplacian_img)

    sobel_vertical=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    sobel_vertical_img=cv2.filter2D(grayscale,-1,sobel_vertical)
    plt.subplot(5,1,4)
    plt.title("Sobel_Vertical Filter",fontsize=150)
    plt.imshow(sobel_vertical_img,cmap="gray")

    sobel_horizontal=np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
    sobel_horizontal_img=cv2.filter2D(grayscale,-1,sobel_horizontal)
    plt.subplot(5,1,5)
    plt.title("Sobel_Horizontal Filter",fontsize=150)
    plt.imshow(sobel_horizontal_img,cmap="gray")

    plt.savefig("Laplacian_&_Sobel_Filters.jpeg")

if __name__=='__main__':
    main()
```
